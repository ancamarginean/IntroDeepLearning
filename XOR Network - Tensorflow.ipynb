```python
%matplotlib inline
import tensorflow as tf
from tensorflow import keras
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
print(tf.__version__)
```

## Learning Objectives:

- Create a neural network for learning XOR - a non-linear separable example.
- Plot the learnt separation surface

```python
x = np.array([[1,1],[1,0],[0,1],[0,0]])
y = np.array([0,1,1,0])
```

## Build the model
- How many layers to use in the model:
> input layer, one hidden layer with 2 neurons, one output layer with 1 neuron.


```python


model = keras.Sequential()
model.add(keras.layers.Dense(2, input_dim=2, activation='sigmoid', 
                              kernel_initializer='he_uniform'))
model.add(keras.layers.Dense(1, activation=tf.nn.sigmoid))
#AdamOptimizer(learning_rate=0.001)
#test GD with lr=0.001 - it fails to completely learn
model.compile(optimizer=tf.train.GradientDescentOptimizer(learning_rate=0.03),
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

## Train the model


```python
history = model.fit(x,
                    y,
                    epochs=10000,
                    batch_size=4,
                    verbose=1)
```

## Evaluate the model


```python
model.summary()
results = model.predict(x)

print(results)
```

```python
X1 = np.arange(-2, 2, 0.25)
Y1 = np.arange(-2, 2, 0.25)
Xm, Ym = np.meshgrid(X1, Y1)

newx_matrix=np.dstack(np.meshgrid(X1,Y1)).reshape(-1,2)
fig = plt.figure(figsize=plt.figaspect(0.5))
ax = fig.add_subplot(1, 1, 1, projection='3d')
y_pred = model.predict(newx_matrix)
Z1=y_pred.reshape(-1, 16)
surf = ax.plot_surface(Xm, Ym, Z1, cmap=cm.coolwarm,linewidth=0, antialiased=False)
ax.set_title('soft threshold separation')

ax.view_init( elev=30, azim=15)
plt.show()
```
