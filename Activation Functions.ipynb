```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
def sigmoid_prime(x):
    return sigmoid(x)*(1-sigmoid(x))
```

```python
sigmoid(0.5), sigmoid(0.52), sigmoid_prime(0.5)
```

```python
def relu(x):
    return np.maximum(0, x)
def step(x):
    return np.array(x > 0, dtype=np.int)
def tanh(x):
    return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
```

```python
X=np.linspace(-5,5,100)
plt.plot(X, sigmoid(X),label='Sigmoid',color='b')
plt.plot(X, tanh(X),label='tanh',color='y')
plt.plot(X, step(X), label='Step', color='g', lw=1, )
plt.plot(X, relu(X), label='ReLU', color='r', lw=1, )


plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Activation Functions')
plt.grid()
plt.ylim(-1.1, 1.1)
plt.legend()
plt.text(4, 0.8, r'$\sigma(x)=\frac{1}{1+e^{-x}}$', fontsize=16)
plt.text(4, 0.4, r'$tanh(x)=\frac{e^{x}-e^{-x}}{e^{x}+e^{-x}}$', fontsize=16)
plt.text(4, -0.25,   r'$step(x)=\{$',fontsize=16)
plt.text(7, -0.15, r'$0,\ x<0$', fontsize=16)
plt.text(7, -0.4, r'$1,\  x \geq 0$', fontsize=16)
plt.text(4, -1,   r'$relu(x)=\{$',fontsize=16)
plt.text(7, -0.95, r'$0,\ x<0$', fontsize=16)
plt.text(7, -1.15, r'$x,\  x \geq 0$', fontsize=16)

plt.show()
```

```python
def relu_prime(x):
    return np.array(x > 0, dtype=np.int) #0, x<0, 1, for x>=0
def step_prime(x):
    return np.array(x > 0, dtype=np.int)
def tanh_prime(x):
    return (1-tanh(x)*tanh(x))
```

```python
X=np.linspace(-5,5,100)

fig, axes = plt.subplots(1, 2, figsize=(12, 8))
ax = axes.ravel()
fig.tight_layout()


ax[0].plot(X, sigmoid(X),label='Sigmoid',color='b')
ax[0].plot(X, tanh(X),label='tanh',color='y')
ax[0].plot(X, relu(X), label='ReLU', color='r', lw=1, )


plt.xlabel('X Axis')
plt.ylabel('Y Axis')
ax[0].set_title('Activation Functions')
ax[0].grid()
ax[0].set_ylim(-1.1, 1.1)

ax[1].plot(X, sigmoid_prime(X),label='Sigmoid',color='b')
ax[1].plot(X, tanh_prime(X),label='tanh',color='y')
ax[1].plot(X, relu_prime(X), label='ReLU', color='r', lw=1, )



ax[1].set_title('Derivative of the Activation Functions')
ax[1].grid()
ax[1].set_ylim(-1.1, 1.1)

ax[1].text(4, 0.8, r'$\sigma^\prime(x)=f(x)(1-f(x)$', fontsize=16)
ax[1].text(4, 0.4, r'$tanh^\prime(x)=1-f^2(x)$', fontsize=16)
ax[1].text(4, -0.25,   r'$relu^\prime(x)=\{$',fontsize=16)
ax[1].text(7, -0.15, r'$0,\ x<0$', fontsize=16)
ax[1].text(7, -0.4, r'$1,\  x \geq 0$', fontsize=16)


plt.legend()
plt.show()
```
